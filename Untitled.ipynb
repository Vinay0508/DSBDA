#### a) Scan all variables for missing values and inconsistencies. If there are missing values and/or inconsistencies, use any of the suitable techniques to deal with them.

```python
df.isnull().sum()
```

```python
numerical_null = [feat for feat in df.columns if df[feat].dtypes != 'O' and df[feat].isnull().sum() > 0]
```

```python
print(numerical_null)
```

```python
categorical_null =  [feat for feat in df.columns if df[feat].dtypes == 'O' and df[feat].isnull().sum() > 0]
```

```python
print(categorical_null)
```

```python
for i in numerical_null:
    df[i] = df[i].replace(np.NaN, df[i].mean())
```

```python
df.isnull().sum()
```

```python
for i in categorical_null:
    print(f"{i} has {np.round(df[i].isnull().mean() * 100,4)}% null values")
```

```python
for i in categorical_null:
    if i != 'EDUCATION_TYPE':
        mode_value = df[i].mode()
        df[i] = np.where(df[i].isnull(), mode_value, df[i])
    else:
        df[i] = np.where(df[i].isnull(), 'Not apply', df[i])
```

```python
for i in categorical_null:
    print(f"{i} has {np.round(df[i].isnull().mean() * 100,4)}% null values")
```

#### b) Scan all numeric variables for outliers. If there are outliers, use any of the suitable techniques to deal with them.

```python
numerical_feat = [feat for feat in df.columns if df[feat].dtypes != 'O']
```

```python
numerical_feat
```

```python
for i in numerical_feat:
    sns.boxplot(x = df[i])
    plt.show()
```

```python
for i in numerical_feat:
    data = sorted(df[i])
    Q1 = np.percentile(data, 25)
    Q3 = np.percentile(data, 75)
    IQR = Q3 - Q1
    low_limit = Q1 - (1.5*IQR)
    up_limit = Q3 + (1.5*IQR)

    mean_value = (up_limit - low_limit)/2
    df[i] = np.where((df[i] < low_limit), low_limit,np.where((df[i]>up_limit),up_limit, df[i]))
```

```python
for i in numerical_feat:
    sns.boxplot(x = df[i])
    plt.show()
```
